```python
import numpy as np
import matplotlib.pyplot as plt

from fieldlearn.drawing import draw_polyvector_field

from PIL import Image
from scipy import ndimage
```

```python
raster = Image.open("/home/mtaktash/data/abc_field_examples/00050009_f5e95c39e7f941fdad8c83e5_step_000_raster.png")
field = np.load("/home/mtaktash/data/abc_field_examples/00050009_f5e95c39e7f941fdad8c83e5_step_000_field.npy")
```

```python
raster
```

```python
raster = np.array(raster)
h_slice = slice(0, 25)
w_slice = slice(0, 35)
raster = raster[h_slice, w_slice]
field  = field[:, h_slice, w_slice]
```

```python
u = field[:2]
v = field[2:]
draw_polyvector_field(u, v, raster, figscale=0.4);
```

```python
from fieldlearn.utils import rotate
```

```python
u_new, v_new, raster_new = rotate(u, v, raster, degrees=20)
draw_polyvector_field(u_new, v_new, raster_new, figscale=0.3);
```
